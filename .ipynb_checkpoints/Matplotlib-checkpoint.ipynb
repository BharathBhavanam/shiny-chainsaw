### Matplotlib

figure-canvas

axes-individual plots inside the figure

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0,10,100)
```

```python
y = np.sin(x)
```

```python
fig,axe=plt.subplots()
axe.plot(x,y,label="sin(x)",color="green")
axe.set_xlabel("x")
axe.set_ylabel("y")
axe.set_title("sine_wave")
axe.legend()
plt.show
```

```python
np.random.seed(42)
```

```python
epochs = np.arange(1,21)
```

```python
train_loss = np.exp(-epochs/5) +np.random.rand(20)*0.5
```

```python
val_loss = np.exp(-epochs/6) +np.random.rand(20)*0.08
```

```python
plt.plot(epochs,train_loss,label='train')
plt.plot(epochs,val_loss,label='validity')
```

### Barplot

```python
models = ["LogReg","RF","SVM","XGB"]
acc = [0.82,0.88,0.85,0.91]
```

```python
plt.bar(models,acc,color="orange")
plt.ylabel("accuracy")
plt.xlabel("models")
plt.title("Model Comparison")
plt.show
```

### Seaborn

```python
iris = sns.load_dataset("iris")
tips = sns.load_dataset("tips")
```

```python
iris.head()
```

```python
sns.histplot(data=iris, x = "sepal_length", hue = "species",kde=True)
```

```python
sns.boxplot(data=iris, x = "species", y = "petal_length")
plt.title("Box-plot")
plt.show()
```

```python
sns.violinplot(data=iris, x= "species", y = "petal_width")
plt.title(["violin plot - Distribution shape"])
plt.show()
```

```python
#heatmap
corr = iris.drop(columns=["species"]).corr()
```
