### MISSING DATA

```python
messy = pd.DataFrame({
    "Product" : ["Pen","Book","Bag","Lamp","Mug"],
    "Price" : [20,np.nan,850,1200,np.nan],
    "Rating" : [4.1,4.5,np.nan,4.5,3.9]
})
```

```python
messy
```

```python
messy.isnull().sum()
```

```python
filled = messy.copy()
```

```python
filled["Price"] = filled["Price"].fillna(filled["Price"].mean())
```

```python
filled["Rating"] = filled["Rating"].fillna(filled["Rating"].mean())
```

```python
print(filled)
```
